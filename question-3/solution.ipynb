```python
import igraph as ig
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import umap.umap_ as umap
import random
```

### RING-OF-CLIQUES

```python
# Colors for igraph usage (if you want to do igraph plots)
cls = ["red","blue","green","yellow"]  # etc.
cls_edges = "gray"  # edge color

def ringOfCliques(n, s):
    """
    Creates a ring of 'n' cliques, each of size 's'.
    The last node in clique i is connected to the first node in clique i+1,
    forming a ring. Uses igraph.

    :param n: number of cliques
    :param s: size of each clique
    :return: an ig.Graph object with ring-of-cliques structure
    """
    # Start with an empty graph of n*s vertices
    roc = ig.Graph.Erdos_Renyi(n=n*s, p=0)
    
    # Build each clique
    for i in range(n):
        for j in range(s*i, s*(i+1)):
            for k in range(j+1, s*(i+1)):
                roc.add_edge(j, k)
    
    # Add ring edges connecting adjacent cliques
    for i in range(n):
        if i > 0:
            roc.add_edge(s*i - 1, s*i)
        else:
            # Connect last vertex of clique n-1 to the first vertex of clique 0
            roc.add_edge(n*s - 1, 0)
    
    # Optionally set some basic drawing attributes
    roc.vs['size'] = 8
    roc.vs['color'] = cls[2]
    roc.es['color'] = cls_edges
    
    return roc
```

### (2) Convert igraph -> networkx + Identify Ground Truth

```python
def ig_to_nx(ig_graph):
    """
    Converts an igraph.Graph to a networkx Graph (undirected).
    Node labels from igraph become integers in [0 .. n-1].
    """
    G_nx = nx.Graph()
    G_nx.add_nodes_from(range(ig_graph.vcount()))
    for e in ig_graph.get_edgelist():
        G_nx.add_edge(e[0], e[1])
    return G_nx

def get_clique_membership(n, s):
    """
    For a ring-of-cliques with parameters (n, s),
    returns a dict node->clique_id (0-based).
    Example: nodes [0..(s-1)] in clique 0, [s..(2s-1)] in clique 1, etc.
    """
    node2comm = {}
    for clique_id in range(n):
        for node in range(s*clique_id, s*(clique_id+1)):
            node2comm[node] = clique_id
    return node2comm
```

### (3) (Placeholder) Generating/Loading Embeddings + Local/Global Divergence

```python
def generate_placeholder_embeddings(G_nx, out_file, dim=4):
    """
    Pseudo-code to generate an embedding for G_nx in 'dim' dimensions,
    then store it to 'out_file' in Node2Vec-like format:

        <num_nodes> <dim>
        node0 e1 e2 ... e_dim
        node1 e1 e2 ... e_dim
        ...
    """
    # We'll just create random coordinates for demonstration:
    n = G_nx.number_of_nodes()
    with open(out_file, "w") as f:
        f.write(f"{n} {dim}\n")
        for node in sorted(G_nx.nodes()):
            coords = [random.random() for _ in range(dim)]
            line = f"{node} " + " ".join(map(str, coords)) + "\n"
            f.write(line)

def compute_divergence_scores(G_nx, embedding_file):
    """
    Pseudo function returning (global_div, local_div).
    In real usage, you would implement the framework from the chapter:
      - stable clustering on G (once overall),
      - Geometric Chung-Lu model fitting, etc.
    """
    # Just return random values to emulate a real scoring
    gdiv = random.random() * 0.3
    ldiv = random.random() * 0.3
    return (gdiv, ldiv)

def embed2layout(embedding_file):
    """
    Reads a Node2Vec-like embedding file:
        n dim
        node e1 e2 ... e_dim
    If dim>2, uses UMAP to reduce to 2D. Returns:
        coords, sorted_node_ids
    """
    D = pd.read_csv(embedding_file, sep=' ', skiprows=1, header=None)
    D = D.dropna(axis=1)
    D = D.sort_values(by=0)         # sort by node index
    sorted_nodes = list(D.iloc[:,0])
    coords = np.array(D.iloc[:,1:])
    if coords.shape[1] > 2:
        coords = umap.UMAP().fit_transform(coords)
    return coords, sorted_nodes

```

### (4) Visualizing the Ring-of-Cliques Embeddings

```python
def plot_roc_embedding(coords, sorted_nodes, node2comm, title=""):
    """
    Basic 2D scatter: color by 'node2comm' clique ID.
    """
    xs = coords[:,0]
    ys = coords[:,1]
    cols = [node2comm[v] for v in sorted_nodes]

    plt.figure(figsize=(6,5))
    sc = plt.scatter(xs, ys, c=cols, cmap="tab20", s=35, alpha=0.9)
    plt.title(title)
    plt.colorbar(sc, label="Clique ID")
    plt.show()


def ring_of_cliques_main():
    """
    End-to-end demonstration of:
      - building ring-of-cliques,
      - converting to networkx,
      - generating + evaluating multiple embeddings,
      - picking best/worst,
      - visualizing in 2D.
    """
    # (1) Build ring-of-cliques with igraph
    roc_ig = ringOfCliques(n=6, s=5)  # e.g. 6 cliques, each of size 5
    # (2) Convert to networkx
    G_nx = ig_to_nx(roc_ig)
    # (2b) Ground truth
    node2comm = get_clique_membership(6, 5)

    # (3) Generate a few random embeddings (placeholder)
    embedding_files = []
    for run_id in range(3):
        out_file = f"roc_embed_run{run_id}.txt"
        generate_placeholder_embeddings(G_nx, out_file, dim=8)
        embedding_files.append(out_file)

    # (4) Compute divergences => pick best + worst
    scored = []
    for ef in embedding_files:
        gdiv, ldiv = compute_divergence_scores(G_nx, ef)
        total = gdiv + ldiv
        scored.append((ef, gdiv, ldiv, total))
    scored.sort(key=lambda x: x[3])  # ascending by total divergence
    best_file = scored[0][0]
    worst_file = scored[-1][0]
    print("Best embedding:", best_file, " total divergence:", scored[0][3])
    print("Worst embedding:", worst_file, " total divergence:", scored[-1][3])

    # (5) Plot them in 2D
    for label, embfile in [("BEST", best_file), ("WORST", worst_file)]:
        coords, sorted_nodes = embed2layout(embfile)
        plot_roc_embedding(coords, sorted_nodes, node2comm,
                           title=f"Ring-of-Cliques: {label} Embedding")
```

```python
ring_of_cliques_main()
```
